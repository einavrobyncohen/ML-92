# Feature selection (with variance)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diabetes_binary = pd.read_csv('processed_diabetes_binary.csv')
```

```python
features = diabetes_binary['Diabetes_binary'].astype(int)
target = diabetes_binary.drop(columns=['Diabetes_binary']).astype(int)
X_train, X_test, y_train, y_test = train_test_split(target, features, test_size=0.2, random_state=92)
```

```python
# Scale data
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
```

```python
fig, ax = plt.subplots()

x = target.columns
y = scaled_X_train.var(axis=0)

ax.bar(x,y, width=0.5)
ax.set_xlabel("Features")
ax.set_ylabel("Variance")
ax.set_ylim(0, 0.5)
#ax.set_xlim(0, 6)

for index, value in enumerate(y):
    plt.text(x=index, y=value+0.001, s=str(round(value, 2)), ha='center')

fig.autofmt_xdate()
plt.tight_layout
# small variance (close to 0) means we don't use that feature
```
