# TomekLinks undersampling method

This notebook will apply the TomekLinks undersampling balancing method to the training data.

```python
import pandas as pd

df_train = pd.read_csv('train.csv')
X_train = df_train.drop('Diabetes_binary', axis=1)
y_train = df_train['Diabetes_binary']

print("Pre-sampling class imbalance:")
print(y_train.value_counts(normalize=True))
```

```python
from imblearn.under_sampling import TomekLinks

tomek = TomekLinks()
X_tomek, y_tomek = TomekLinks.fit_resample(X_train, y_train)

print("Class distribution after TomekLinks:")
print(y_tomek.value_counts(normalize=True))
```
