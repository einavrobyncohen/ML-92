```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sb
```

- 0 = no diabetes 
- 1 = prediabetes  
- 2 = diabetes

```python
# Read the csv and store in variable diabetes_data
diabetes_data = pd.read_csv("diabetes_multiclass.csv") 
# Filter the data based on the diabetes categories (no diabetes, prediabetic, diabetes)
diabetes_0 = diabetes_data[diabetes_data['Diabetes_012'] == 0]
diabetes_1 = diabetes_data[diabetes_data['Diabetes_012'] == 1]
diabetes_2 = diabetes_data[diabetes_data['Diabetes_012'] == 2]
```

# Features: Diabetes_012 vs Education
Education level (EDUCA see codebook) scale 1-6 
- 1 = Never attended school or only kindergarten 
- 2 = Grades 1 through 8 (Elementary) 
- 3 = Grades 9 through 11 (Some high school) 
- 4 = Grade 12 or GED (High school graduate) 
- 5 = College 1 year to 3 years (Some college or technical school) 
- 6 = College 4 years or more (College graduate)

```python
# Count occurrences of each education level wrt diabetes and sort from 1 to 6
education0_counts = diabetes_0['Education'].value_counts().sort_index()
education1_counts = diabetes_1['Education'].value_counts().sort_index()
education2_counts = diabetes_2['Education'].value_counts().sort_index()

# Size of figure
plt.figure(figsize=(8, 6.5))

# Plotting the bars
x = 1 + np.arange(6)
plt.bar(x, education0_counts, width = 0.2, color='plum', label='non diabetic')
plt.bar(x+0.2, education1_counts, width = 0.2, color='salmon', label='pre-diabetic')
plt.bar(x-0.2, education2_counts, width = 0.2, color='lightskyblue', label='diabetic')

# Labels + title
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.title('Frequency of Education Levels for Diabetes_012')
# Set values on x-axis horizontal
plt.xticks(rotation=0)
# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
