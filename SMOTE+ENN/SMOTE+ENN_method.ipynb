# SMOTE+ENN Method

This notebook will apply the SMOTE+ENN balancing method to the training data.

```python
import pandas as pd

df_train = pd.read_csv('train.csv')
X_train = df_train.drop('Diabetes_binary', axis=1)
y_train = df_train['Diabetes_binary']

print("Pre-sampling class imbalance:")
print(y_train.value_counts(normalize=True))
```

```python
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN(random_state=92)
X_smoteenn, y_smoteen = smote_enn.fit_resample(X_train, y_train)

print("Class distribution after SMOTE+ENN:")
print(y_smoteen.value_counts(normalize=True))
```
